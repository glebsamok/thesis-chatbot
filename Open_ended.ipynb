### Aggregate the scores for traits

```python
def replace_likert_values(value):
  if value == "Strongly agree":
    return 3
  elif value == "Agree":
    return 2
  elif value == "Somewhat agree":
    return 1
  elif value == "Neither agree nor disagree":
    return 0
  elif value == "Somewhat disagree":
    return -1
  elif value == "Disagree":
    return -2
  elif value == "Strongly disagree":
    return -3
  else:
    return value

emotional_intelligence_dataset = emotional_intelligence_dataset.applymap(replace_likert_values)
perspective_taking_dataset = perspective_taking_dataset.applymap(replace_likert_values)
learning_orientation_dataset = learning_orientation_dataset.applymap(replace_likert_values)
social_curiosity_dataset = social_curiosity_dataset.applymap(replace_likert_values)
big5_dataset = big5_dataset.applymap(replace_likert_values)
```

```python
def calculate_row_sum_filtered(df):
  df_numeric = df.apply(pd.to_numeric, errors='coerce')

  row_sums = []

  for index, row in df_numeric.iterrows():
    current_row_sum = 0
    for value in row:
      if pd.notna(value) and -4 <= value <= 4:
        current_row_sum += value
    row_sums.append(current_row_sum)

  df['agg_sum'] = row_sums
  return df

emotional_intelligence_dataset = calculate_row_sum_filtered(emotional_intelligence_dataset)
perspective_taking_dataset = calculate_row_sum_filtered(perspective_taking_dataset)
learning_orientation_dataset = calculate_row_sum_filtered(learning_orientation_dataset)
social_curiosity_dataset = calculate_row_sum_filtered(social_curiosity_dataset)
big5_dataset = calculate_row_sum_filtered(big5_dataset)
```

### Categorize the aggregate scores into labels for the LLM to predict

```python
def categorize_trait(score_series, trait_name):
    mean = score_series.mean()
    std = score_series.std()

    def label(score):
        if score >= mean + 1 * std:
            return f"High"
        elif score <= mean - 1 * std:
            return f"Low"
        else:
            return f"Medium"

    return score_series.apply(label)
```

```python
# Apply to your DataFrame
merged_data['emotional_intelligence_label'] = categorize_trait(merged_data['emotional_intelligence_score'], "Emotional Intelligence")
merged_data['perspective_taking_label'] = categorize_trait(merged_data['perspective_taking_score'], "Perspective Taking")
merged_data['learning_orientation_label'] = categorize_trait(merged_data['learning_orientation_score'], "Learning Orientation")
merged_data['social_curiosity_label'] = categorize_trait(merged_data['social_curiosity_score'], "Social Curiosity")
merged_data['big5_label'] = categorize_trait(merged_data['big5_score'], "Big Five")
```

```python
def generate_full_soft_skills_prompt_0(row):
    # Introductory instructions and context
    header = (
        "You are a psychologist trained in evaluating soft skills based on people's answers to open-ended questions.\n\n"
        "Now evaluate the following person's emotional intelligence, perspective taking, learning orientation and curiosity based on their answers to the following questions.\n\n"
    )

    # First question block
    intro_1 = (
        "Think about a frustrating moment you had with a leader (e.g., your direct manager, "
        "a project lead) or a colleague (e.g., a team member, someone from another department) "
        "in the past two weeks. It could be a disagreement, misunderstanding, or just a situation "
        "that left you feeling annoyed. Please answer the following questions about that specific situation:\n\n"
    )

    questions_1 = [
        'Briefly describe what happened (1-2 sentences).',
        'How did this situation make you feel? How did you express your feelings? Did you avoid them, expressed them, or kept them to yourself?',
        'What emotions do you think the other person was feeling? What cues gave you this impression? Why those behaviours indicated the following emotions?',
        'Did you understand where they were coming from? What do you think was driving their behavior?',
        'Looking, deeper, what underlying needs or concerns might have influenced how they acted?',
    ]

    # Second question block
    intro_2 = (
        "\n\nThink about something you've been working toward in the past month - maybe aiming for a promotion, "
        "learning a new job skill, managing a tough project, or getting a certification. What got you interested "
        "in pursuing this goal in the first place? Please answer the following questions about that specific goal:\n\n"
    )

    questions_2 = [
        "Briefly describe your goal and what you've been doing to achieve it (1-2 sentences).",
        'When looking at all the information you gathered while working on this goal, was there a time when things felt particularly confusing or complicated? How did you sort through that confusion to better understand what you needed to do?',
        'What unexpected challenges or surprises have you faced while working toward this goal? What steps did you take to overcome them?',
        'Did you try any new approaches or methods while working on this goal? Where did you learn about this new approach and how did you put it into practice?'
    ]

    # Build Part 1 content
    part_1 = "\n".join([f"{q}: {row[q]}" for q in questions_1 if pd.notnull(row[q])])
    part_2 = "\n".join([f"{q}: {row[q]}" for q in questions_2 if pd.notnull(row[q])])

    # Final evaluation request
    closing = (
        "\n\nBased only on the above answers, how would you rate this person's emotional intelligence, "
        "learning orientation, perspective taking and curiosity?\n\n"
        "Please respond with one label of: **Low**, **Medium**, or **High** in the format of {'Emotional Intelligence': 'Label', 'Perspective Taking':'Label', 'Learning Orientation': 'Label', 'Curiosity':'Label'}"
    )

    return header + intro_1 + part_1 + intro_2 + part_2 + closing
```

```python
import openai
import os
```

```python
# Create a client
client = openai.OpenAI(api_key="") #api

# Function to get soft skill ratings from the LLM
def get_soft_skill_labels(prompt):
    response = client.chat.completions.create(
        model="gpt-4o-mini",  # or "gpt-4o"
        messages=[
            {"role": "user", "content": prompt}
        ],
        temperature=0
    )
    return response.choices[0].message.content
```

```python
final_data['prompt_0'] = final_data.apply(generate_full_soft_skills_prompt_0, axis=1)
```

```python
final_data['LLM_labels_0'] = final_data['prompt_0'].apply(get_soft_skill_labels)
```

```python
import ast  # safe way to evaluate string representations of Python literals

# Safely convert the string to a dictionary
def parse_llm_output(output_str):
    try:
        return ast.literal_eval(output_str)
    except Exception as e:
        print(f"Error parsing: {output_str}\n{e}")
        return {}

# Apply parsing to your column
parsed_0 = final_data['LLM_labels_0'].apply(parse_llm_output)

# Extract each trait into its own column
final_data['emotional_intelligence_label_llm_0'] = parsed_0.apply(lambda x: x.get('Emotional Intelligence'))
final_data['perspective_taking_label_llm_0'] = parsed_0.apply(lambda x: x.get('Perspective Taking'))
final_data['learning_orientation_label_llm_0'] = parsed_0.apply(lambda x: x.get('Learning Orientation'))
final_data['curiosity_label_llm_0'] = parsed_0.apply(lambda x: x.get('Curiosity'))
```

```python
# Split final_dataset into four â€œtruth / predictedâ€ DataFrames
emotional_intelligence_df_few_shot_0 = final_data[
    ["emotional_intelligence_label", "emotional_intelligence_label_llm_0"]
].copy()

perspective_taking_df_few_shot_0 = final_data[
    ["perspective_taking_label", "perspective_taking_label_llm_0"]
].copy()

learning_orientation_df_few_shot_0 = final_data[
    ["learning_orientation_label", "learning_orientation_label_llm_0"]
].copy()

social_curiosity_df_few_shot_0 = final_data[
    ["social_curiosity_label", "curiosity_label_llm_0"]
].copy()

# Rename for consistency (optionalâ€”labels already OK)
for df in [
    emotional_intelligence_df_few_shot_0,
    perspective_taking_df_few_shot_0,
    learning_orientation_df_few_shot_0,
    social_curiosity_df_few_shot_0,
]:
    df.columns = ["truth", "predicted"]

# Quick sanity-check
print(emotional_intelligence_df_few_shot_0.head(5))
```

```python
import pandas as pd
from sklearn.metrics import (
    accuracy_score,
    precision_recall_fscore_support,
    confusion_matrix,
)

# ------------------------------------------------------------------
# 1) Re-use / (re)define the helper for our 3-class text labels
# ------------------------------------------------------------------
def evaluate_low_med_high(df, y_true="truth", y_pred="predicted",
                          label_order=["Low", "Medium", "High"],
                          show_report=False):
    """Return accuracy, macroâ€“ & weightedâ€“averages, per-class table & CM."""
    y_t, y_p = df[y_true], df[y_pred]

    acc = accuracy_score(y_t, y_p)

    p_macro, r_macro, f1_macro, _ = precision_recall_fscore_support(
        y_t, y_p, average="macro", zero_division=0
    )
    p_wgt, r_wgt, f1_wgt, _ = precision_recall_fscore_support(
        y_t, y_p, average="weighted", zero_division=0
    )

    p, r, f1, support = precision_recall_fscore_support(
        y_t, y_p, labels=label_order, average=None, zero_division=0
    )
    per_class = pd.DataFrame(
        {"precision": p, "recall": r, "f1": f1, "support": support},
        index=label_order,
    )

    cm = confusion_matrix(y_t, y_p, labels=label_order)

    if show_report:
        # Pretty one-liner if you still want the full string report
        from sklearn.metrics import classification_report
        print(classification_report(y_t, y_p, labels=label_order, zero_division=0))

    return {
        "accuracy": acc,
        "macro": {"precision": p_macro, "recall": r_macro, "f1": f1_macro},
        "weighted": {"precision": p_wgt, "recall": r_wgt, "f1": f1_wgt},
        "per_class": per_class,
        "confusion_matrix": cm,
    }

# ------------------------------------------------------------------
# 2) Bundle the four DataFrames and loop
# ------------------------------------------------------------------
trait_frames = {
    "Emotional Intelligence": emotional_intelligence_df_few_shot_0,
    "Perspective Taking":     perspective_taking_df_few_shot_0,
    "Learning Orientation":   learning_orientation_df_few_shot_0,
    "Social Curiosity":       social_curiosity_df_few_shot_0,
}

summary_rows = []   # collect headline numbers
matrices = {}       # store CMs for later plotting if you like

for name, df in trait_frames.items():
    res = evaluate_low_med_high(df, show_report=False)
    summary_rows.append({
        "Trait": name,
        "Accuracy":  res["accuracy"],
        "Macro F1":  res["macro"]["f1"],
        "Weighted F1": res["weighted"]["f1"],
    })
    matrices[name] = res["confusion_matrix"]

summary_df = pd.DataFrame(summary_rows).set_index("Trait")
print("Headline metrics per trait:\n")
print(summary_df.round(3))

# ------------------------------------------------------------------
# 3) (Optional) inspect confusion matrices
# ------------------------------------------------------------------
for name, cm in matrices.items():
    print(f"\n{name} â€“ confusion matrix (rows = truth, cols = pred):")
    print(pd.DataFrame(cm,
                       index=["Low","Medium","High"],
                       columns=["Low","Medium","High"]))
```

```python
import pandas as pd
from sklearn.metrics import (
    accuracy_score,
    precision_recall_fscore_support,
    confusion_matrix,
)

# ------------------------------------------------------------------
# 1) Re-use / (re)define the helper for our 3-class text labels
# ------------------------------------------------------------------
def evaluate_low_med_high(df, y_true="truth", y_pred="predicted",
                          label_order=["Low", "Medium", "High"],
                          show_report=False):
    """Return accuracy, macroâ€“ & weightedâ€“averages, per-class table & CM."""
    y_t, y_p = df[y_true], df[y_pred]

    acc = accuracy_score(y_t, y_p)

    p_macro, r_macro, f1_macro, _ = precision_recall_fscore_support(
        y_t, y_p, average="macro", zero_division=0
    )
    p_wgt, r_wgt, f1_wgt, _ = precision_recall_fscore_support(
        y_t, y_p, average="weighted", zero_division=0
    )

    p, r, f1, support = precision_recall_fscore_support(
        y_t, y_p, labels=label_order, average=None, zero_division=0
    )
    per_class = pd.DataFrame(
        {"precision": p, "recall": r, "f1": f1, "support": support},
        index=label_order,
    )

    cm = confusion_matrix(y_t, y_p, labels=label_order)

    if show_report:
        # Pretty one-liner if you still want the full string report
        from sklearn.metrics import classification_report
        print(classification_report(y_t, y_p, labels=label_order, zero_division=0))

    return {
        "accuracy": acc,
        "macro": {"precision": p_macro, "recall": r_macro, "f1": f1_macro},
        "weighted": {"precision": p_wgt, "recall": r_wgt, "f1": f1_wgt},
        "per_class": per_class,
        "confusion_matrix": cm,
    }

# ------------------------------------------------------------------
# 2) Bundle the four DataFrames and loop
# ------------------------------------------------------------------
trait_frames = {
    "Emotional Intelligence": emotional_intelligence_df_few_shot_1,
    "Perspective Taking":     perspective_taking_df_few_shot_1,
    "Learning Orientation":   learning_orientation_df_few_shot_1,
    "Social Curiosity":       social_curiosity_df_few_shot_1,
}

summary_rows = []   # collect headline numbers
matrices = {}       # store CMs for later plotting if you like

for name, df in trait_frames.items():
    res = evaluate_low_med_high(df, show_report=False)
    summary_rows.append({
        "Trait": name,
        "Accuracy":  res["accuracy"],
        "Macro F1":  res["macro"]["f1"],
        "Weighted F1": res["weighted"]["f1"],
    })
    matrices[name] = res["confusion_matrix"]

summary_df = pd.DataFrame(summary_rows).set_index("Trait")
print("Headline metrics per trait:\n")
print(summary_df.round(3))

# ------------------------------------------------------------------
# 3) (Optional) inspect confusion matrices
# ------------------------------------------------------------------
for name, cm in matrices.items():
    print(f"\n{name} â€“ confusion matrix (rows = truth, cols = pred):")
    print(pd.DataFrame(cm,
                       index=["Low","Medium","High"],
                       columns=["Low","Medium","High"]))
```
